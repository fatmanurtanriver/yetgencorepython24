# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
# 1.Problemin Çözümünü Buraya Yazınız
import time

def girdi():
    while True:
        try:
            enter=int(input("kaç isim girişi yapacaksınız ?"))
            if enter<=0:
                print("lütfen pozitif bir sayı giriniz.")
            else:
                return enter
        except ValueError:
            print("lütfen bir sayı giriniz.")
            

kisi_sayisi=girdi()
isimler=[]
yaslar=[]
boylar=[]
kilolar=[]

for i in range(kisi_sayisi):
    isim=input("isim-soyisim:")
    yas=int(input("yaş:"))
    boy=float(input("boy(cm):"))
    kilo=float(input("kilo(kg):"))
    isimler.append(isim)
    yaslar.append(yas)
    boylar.append(boy)
    kilolar.append(kilo)

def vki_hesap(isimler,yaslar,boylar,kilolar):
    with open ("vki listesi2.txt", "a", encoding="utf-8") as file:
        for a in range(len(isimler)):
            isim=isimler[a]
            yas=yaslar[a]
            boy=boylar[a]
            kilo=kilolar[a]

            boy_m=boy/100
            vki=kilo/(boy_m**2)
            vki=round(vki,1)

            if vki<18.5:
                sinif="zayıf"
            elif 18.5<= vki <= 24.9:
                sinif="normal kilolu"
            elif 25<= vki <=29.9:
                sinif="fazla kilolu"
            elif 30<= vki <=39.9:
                sinif="obez"
            elif 40<=vki:
                sinif="ileri derecede (morbid) obez"
            
            file.write(f"{isim}:\nyaş:{yas}\nboy:{boy} cm\nkilo:{kilo} kg\nvki={vki} kg/m\ndurum:{sinif}\n\n")
            file.write(f"{time.ctime(time.time())}\n")
            file.write("-"*15)
            file.write("\n")
            print("kullanıcı kaydedildi.")
        

vki_hesap(isimler,yaslar,boylar,kilolar)


    
```

```python
# 2.Problemin Çözümünü Buraya Yazınız (-kisisselsm işlemleri -menu goruntuleme islemleri -satın alma islemleri -)

import time
menu={"muz":5,"ıspanak":12,"protein tozu":15,"inek sütü":6,"badem sütü":8, "yoğurt":5.5, "çilek":10,"chia":4.75 , "yeşil elma":7,"zencefil":9,"ananas":10, "portakal":15,"limon":4, "şeftali":9,"kayısı":8,"karpuz":8,"nane":7,"fıstık ezmesi":10,"kakao":5,"ahududu":14,"böğürtlen":13,"tarçın":5}


def menu_goruntule():
    print("-----YETBAR MENÜ-----")
    for item in menu.keys():
        print(f"{item.upper():<10}:{menu[item]} TL")
        


def urun_hesaplama(urun,adet):
    assert adet>0, "Ürün adedi 0'dan küçük veya 0'a eşit olamaz." 
    if urun not in menu.keys():
        return None
    else:
        return menu[urun]*adet


def satin_alma():

    menu_goruntule()
    print("Lütfen smoothienize eklemek istediğiniz ürünün adını giriniz. Çıkış yapmak için 0'ı tuşlayınız.")
    son_fiyat=0
    satin_alinan_urun=[]
    urun_adedi=[]
    while True:
        urun_bilgisi=input("Hangi ürünü eklemek istersiniz ?")
        if urun_bilgisi=="0":
            break

        ekleme_adedi=int(input("üründen kaç adet eklemek istersiniz ?"))
        fiyat=urun_hesaplama(urun_bilgisi,ekleme_adedi)
        if fiyat is None:
            print("Yanlış değer girdiniz. Ürün kaydedilemedi.")

        else:
            print(f"Toplam ücret {fiyat}.\nSatın almayı onaylamak için y'yi, iptal etmek için n'yi tuşlayınız.")
            check=input("Satın alma işlemini onaylıyor musunuz ?")

            if check=="y":
                print("Satın alma işlemi onaylandı.")
                print("-"*45)
                son_fiyat +=fiyat
                satin_alinan_urun.append(urun_bilgisi)
                urun_adedi.append(ekleme_adedi)
                print(f"Ara Toplam: {son_fiyat}")

            elif check=="n":
                print("Satın alma işlemi iptal edildi.")

    return son_fiyat, satin_alinan_urun,urun_adedi
        
def main():
    print("YetBar Smoothie Bank'e Hoş Geldiniz ! Hangi işlemi yapmak istersiniz ?\nMenüyü görüntülemek için 1'i\nSmoothienizi yapmak için 2'yi\nÇıkış yapmak için 0'ı seçiniz..")
    karar=None
    while True:
        karar=int(input("Hangi işlemi yapmak istiyorsunuz ?"))
        if karar<0 or karar>2:
            print("Hatalı işlem tercihi. Lütfen tekrar deneyiniz.")
        elif karar==0:
            print("Çıkış yapılıyor...\nSağlıklı günler dileriz.")
        elif karar==1:
            menu_goruntule()
        elif karar==2:
            son_fiyat, satin_alinan_urun, urun_adedi =satin_alma()    
            with open ("kasa.txt", "a", encoding="utf-8") as file:
                file.write(time.ctime(time.time()))
                file.write("\n")
                file.write(",".join(map(str,satin_alinan_urun)))
                file.write("\n")
                file.write(",".join(map(str,urun_adedi)))
                file.write("\n")
                file.write(f"Toplam:{son_fiyat}")
                file.write("\n")
                file.write("-"*15)
                file.write("\n")
                
            cikis_kontrol=input("Başka bir işlem yapmak istiyor musunuz ? (Evet/Hayır)")
            if cikis_kontrol=="evet" or cikis_kontrol=="Evet":
                    cikis_kontrol.upper()
                    pass
            else:
                print("\nTeşekkürler, iyi günler dileriz.")
                break

main()     



        

```
