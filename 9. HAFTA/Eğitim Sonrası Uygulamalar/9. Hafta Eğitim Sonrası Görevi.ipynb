Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız

import random
import time

kanallar={"kanal1":2, "kanal2":4,"kanal3":7,"kanal4":10,"kanal5":15,"kanal6":17,"kanal7":20}

class Televizyon():
    def __init__ (self, televizyon_durum="kapalı",ses_seviyesi=10, kanal="kanal1"):
        self.televizyon_durum=televizyon_durum
        self.ses_seviyesi=ses_seviyesi
        self.kanal=kanal

    def tv_ac(self):
        if self.televizyon_durum=="kapalı":
            print("açılıyor...")
            self.televizyon_durum="açık"
        else:
            print("önce televizyonu kapatınız.")

    def tv_kapat(self):
        if self.televizyon_durum=="açık":
            print("televizyon kapatılıyor...")
            self.televizyon_durum="kapalı"
        else:
            print("önce televizyonu açınız.")

    def ses_ayarla(self, yeni_seviye=None):
        if self.televizyon_durum=="kapalı":
            print("önce televizyonu açınız.")
        else:
            pass
        try:
            if 0<=self.ses_seviyesi<=100:
                print(f"ses seviyesi:{self.ses_seviyesi}.")
                yeni_seviye=int(input("istediğiniz ses seviyesini girin:"))
                print("ses seviyesi ayarlanıyor...")
                time.sleep(1)
                self.ses_seviyesi=yeni_seviye
                print(f"ses seviyesi {self.ses_seviyesi} olarak güncellendi.")
        except ValueError:
            print("tam sayı değeri giriniz.")
        except TypeError:
            print("sayısal bir değer giriniz.")
        except yeni_seviye<0 or yeni_seviye>100:
            print ("ses seviyesi 0'dan küçük, 100'den büyük olamaz.")

    def kanal_ekle(self):
        if self.televizyon_durum=="kapalı":
            print("önce televizyonu açınız.")
        else:
            print("----KANAL LİSTESİ-----")
            for item in list(kanallar.keys()):
                print(f"{item.upper()} : {kanallar[item]}")
            while True:
                yeni_kanal=input("eklemek istediğiniz kanal için bir isim belirleyiniz:")
                if yeni_kanal in kanallar.keys():
                    print("halihazırda bu isimde bir kanal ismi mevcut. lütfen başka bir isim seçiniz.")
                else:
                    break
            while True:
                try:
                    kanal_no=int(input("kanalın hangi numarada olmasını istersiniz ?:"))
                    if kanal_no in kanallar.values():
                        print("bu numarada bir kanal mevcut. başka bir numara atayınız.")
                    else:
                        print("kanal ekleniyor...")
                        time.sleep(1)
                        kanallar[yeni_kanal]=[kanal_no]
                        print("kanal eklendi !")
                        break
                except ValueError:
                    print("hatalı işlem. lütfen bir sayı giriniz.")

    def kanal_sayisi_ogren(self):
        if self.televizyon_durum=="kapalı":
            print("önce televizyonu açınız.")
        else:
            kanal=len(kanallar)
            print(f"mevcut kanal sayısı:{kanal}")
        
    def random_kanal_ac(self):
        if self.televizyon_durum=="kapalı":
            print("önce televizyonu açınız.")
        random_kanal=random.choice(list(kanallar.keys()))
        self.kanal=random_kanal
        print(f"rastgele açılan kanal:{random_kanal}")

    def televizyon_bilgileri(self):
        if self.televizyon_durum=="kapalı":
            print("önce televizyonu açınız.")
        else:
            print(f"""
                  Televizyon Bilgisi:
                    izlemekte olduğunuz kanal: {self.kanal}
                    Kanal Numarası: {kanallar[self.kanal]}
                    Televizyon ses seviyesi: {self.ses_seviyesi}
                """)
                

kumanda=Televizyon()
print("""
          İşlemler:
          1. Tv Aç
          2. Tv Kapat
          3. Ses Ayarları
          4. Kanal Ekle
          5. Kanal Sayısını Öğrenme
          6. Rastgele Kanala Geçme
          7. Televizyon Bilgilerini Göster""")

while True:
        islem=input("yapmak istediğiniz işlemin numarasını giriniz:(işlem yapmak istemezeniz q'ya basınız.)")
        if islem=="q":
            break
        elif islem=="1":
            kumanda.tv_ac()
        elif islem=="2":
            kumanda.tv_kapat()
        elif islem=="3":
            kumanda.ses_ayarla()
        elif islem=="4":
            kumanda.kanal_ekle()
        elif islem=="5":
            kumanda.kanal_sayisi_ogren()
        elif islem=="6":
            kumanda.random_kanal_ac()
        elif islem=="7":
            kumanda.televizyon_bilgileri()
        else:
            print("geçersiz işlem.")
               
                  




        
```
